```python
!pip install googletrans==4.0.0-rc1 ipywidgets

from googletrans import Translator
import ipywidgets as widgets
from IPython.display import display

# Inicializa o tradutor
translator = Translator()

# Widgets de entrada
idiomas = {
    'Português': 'pt',
    'Inglês': 'en',
    'Espanhol': 'es',
    'Francês': 'fr',
    'Alemão': 'de',
    'Italiano': 'it',
    'Japonês': 'ja',
    'Chinês': 'zh-cn',
    'Russo': 'ru'
}

origem = widgets.Dropdown(
    options=idiomas,
    value='pt',
    description='De:',
)

destino = widgets.Dropdown(
    options=idiomas,
    value='en',
    description='Para:',
)

texto = widgets.Textarea(
    value='Olá, mundo! Como você está hoje?',
    description='Texto:',
    layout=widgets.Layout(width='100%', height='80px')
)

botao = widgets.Button(description='Traduzir', button_style='success')
saida = widgets.Output()

# Função de tradução
def traduzir_texto(b):
    with saida:
        saida.clear_output()
        try:
            traducao = translator.translate(texto.value, src=origem.value, dest=destino.value)
            print(f'Texto original ({origem.label}): {texto.value}')
            print(f'Tradução ({destino.label}): {traducao.text}')
        except Exception as e:
            print("Erro na tradução:", e)

botao.on_click(traduzir_texto)

# Exibe a interface
display(origem, destino, texto, botao, saida)


```
